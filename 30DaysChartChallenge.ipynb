## Day 1: Comparisons, Part-to-Whole

Donut charts are used to display categorical data proportions, with the size of each component signifying the proportion of each category. This donut chart was made with #matplotlib (Python) and Adobe Photoshop to show the percentage of the population that has been fully vaccinated against COVID-19.

### Data:
* [DOH Philippines VACCINE ROLLOUT UPDATE: 26 July 2021](https://twitter.com/DOHgovph/status/1419536773701464064/photo/1)

```python
# Import Libraries
import matplotlib.pyplot as plt

# Create Data
name = ['1st Dose', 'Fully Vaccinated', 'Unvaccinated']
values = [11113107, 6089314, 110881756]

fig, ax = plt.subplots(figsize=(15,15))

# Create a circle at the center of the plot
my_circle = plt.Circle( (0,0), 0.7, color='#FFE194')

# Give color names
ax.pie(values, labels=name, colors=['#E8F6EF','#B8DFD8','#4C4C6D'], shadow=True, 
       textprops={'fontsize': 20}, autopct='%1.1f%%')
fig = plt.gcf()
fig.gca().add_artist(my_circle)
fig.set_facecolor('#FFE194')
fig.savefig("images/day1.jpeg", facecolor=fig.get_facecolor(), edgecolor='none', dpi=100)

# Show graph
plt.axis("equal")
plt.tight_layout()
plt.show()
```

## Day 2: Comparisons, Pictogram

Pictogram Charts use icons to provide a more visually appealing overall perspective of small collections of discrete data. Here's a pictogram chart I made with PyWaffle. Here is a breakdown of registered voters for the 2019 National Elections by age group; young Filipinos have a good possibility of unseating bad leaders and electing good ones.

### Data: Records and Statistics Division, COMELEC

```python
from pywaffle import Waffle
```

```python
data = {'18-29':18847230, '30-39':14071109, '40-49':11118025, '50-59': 8665320, '60+':9142066}
values = [round(i/1000000, 2) for i in data.values()]

data_final = {}
for i, j in zip(data.keys(), values):
    data_final[i] = j
    
count = 0
for i in data_final.values():
    x = round(i/sum(data_final.values()) * 100,1)
    print(x)
    count += x
    
count
```

```python
fig = plt.figure(
    FigureClass=Waffle, 
    columns=20, 
    values=data_final,
    icons='user', 
    font_size=40, 
    block_arranging_style='new-line',
    vertical=True,
    figsize=(20,20),
    cmap_name='Set2',
    dpi=100,
    facecolor='#515E63',
    legend={
        'labels' : [f"{k} ({int(v / sum(data_final.values()) * 100)}%)" for k, v in data_final.items()],
        'loc': 'lower left',
        'bbox_to_anchor': (0, -0.3),
        'ncol': len(data),
        'framealpha': 0,
        'fontsize': 15
    },
    icon_legend=True)
plt.savefig("images/day2.jpeg")
plt.show()
```
